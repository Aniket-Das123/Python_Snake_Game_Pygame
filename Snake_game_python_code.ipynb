```python
import pygame,random,sys
from pygame.locals import *

pygame.init()

screen_width,screen_height=600,600
grid_size=20
white=(255,255,255)
green=(0,255,0)
red=(255,0,0)
black=(0,0,0)
bg_color=(30,30,30)

screen=pygame.display.set_mode((screen_width,screen_height))
pygame.display.set_caption("Snake Game")
font=pygame.font.SysFont("Arial",30,bold=True)

def draw_grid():
    for x in range(0,screen_width,grid_size):
        pygame.draw.line(screen,(50,50,50),(x,0),(x,screen_height))
    for y in range(0,screen_height,grid_size):
        pygame.draw.line(screen,(50,50,50),(0,y),(screen_width,y))

def collide(x1,y1,x2,y2,size):
    return x1==x2 and y1==y2
def game_over_screen(score):
    screen.fill(bg_color)
    text=font.render(f"Game Over !! score : {score}",True,white)
    restart_text=font.render("Press R to Restart or Q to Quit",True,white)
    screen.blit(text,(screen_width//4,screen_height//3))
    screen.blit(restart_text,(screen_width//6,screen_height//2))
    pygame.display.update()
    while True:
        for e in pygame.event.get():
            if e.type==QUIT or (e.type==KEYDOWN and e.key==K_q):
                pygame.quit()
                sys.exit()
            if e.type==KEYDOWN and e.key==K_r:
                return True

def main():
    speed=6
    clock=pygame.time.Clock()
    snake=[(300,300),(280,300),(260,300)]
    dirs=(grid_size,0)
    apple=(random.randrange(0,screen_width,grid_size),random.randrange(0,screen_height,grid_size))
    score=0
    running=True
    while running:
        screen.fill(bg_color)
        draw_grid()
        for e in pygame.event.get():
            if e.type==QUIT:
                pygame.quit()
                sys.exit()
            elif e.type==KEYDOWN:
                if e.key==K_UP and dirs!=(0,grid_size):
                    dirs=(0,-grid_size)
                elif e.key==K_DOWN and dirs!=(0,-grid_size):
                    dirs=(0,grid_size)
                elif e.key==K_LEFT and dirs!=(grid_size,0):
                    dirs=(-grid_size,0)
                elif e.key==K_RIGHT and dirs!=(-grid_size,0):
                    dirs=(grid_size,0)
        new_head=(snake[0][0]+dirs[0],snake[0][1]+dirs[1])
        if new_head[0]<0 or new_head[0]>=screen_width or new_head[1]<0 or new_head[1]>=screen_height:
            if game_over_screen(score):
                main()
            return
        if new_head in snake:
            if game_over_screen(score):
                main()
            return
        snake.insert(0,new_head)
        if collide(new_head[0],new_head[1],apple[0],apple[1],grid_size):
            score+=1
            apple=(random.randrange(0,screen_width,grid_size),random.randrange(0,screen_height,grid_size))
            speed+=1
        else:
            snake.pop()
        
        for segment in snake:
            pygame.draw.rect(screen,red,(segment[0],segment[1],grid_size,grid_size))
        pygame.draw.rect(screen,green,(apple[0],apple[1],grid_size,grid_size))
        
        score_text=font.render(f"Score : {score}",True,white)
        screen.blit(score_text,(10,10))
        pygame.display.update()
        clock.tick(speed)
        
if __name__ == "__main__":
    main()
```
